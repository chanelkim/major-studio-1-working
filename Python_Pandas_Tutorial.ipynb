# Part 1: Introducing Pandas

Pandas is a Python library for data wrangling and analysis.

* Data structure called the DataFrame - modeled after the R DataFrame
* Simply, it is like a table, similar to an Excel spreadsheet
* Great range of methods to modify and operate on data tables
* SQL-like queries, joins of tables
* Ingest from a variety of data file formats and data bases (SQL, Excel file, csv, json)

#### Import pandas module

## 1. Create a pandas dataframe

```python
# Create a simple dataset using dictionary
data = {'Name': ["John", "Anna", "Peter", "Sarah", "Nick"],
       'Location': ["New York", "Paris", "Berlin", "London", "Barcelona"],
       'Major': ["Graphic Design", "Data Visualization", "Design Technology", "Fashion Design", "Data Science"],
       'Age': [28, 30, 27, 25, 26]}

print (data)
```

```python
# Convert this dictionary to pandas dataframe
data_example = pd.DataFrame(data)
```

```python
# Pandas dataframe allows pretty printing of dataframes
data_example
```

## 2. Data shape, columns, and rows

```python
print ("Shape of data_pandas: ",data_example.shape)

# Print all columns
print ("Column names: ",data_example.columns.tolist())
```

```python
# Call data associated a columne
data_example['Name']
```

```python
data_example.Name
```

```python
# multiple columns

data_example[['Name', 'Major']]
```

## 3. Data types and data selection

```python
data_example.dtypes
```

```python
# Show the rows based on index
data_example.iloc[1:3] # rows with index 1 through 3 (stop before index 3)
```

```python
# Show the rows based on the value
# Syntax: DataFrameName[ConditionStatement]
data_example[data_example['Age']>27]
```

```python
data_example['Age']>27
```

```python
data_example[data_example['Major']=='Graphic Design']
```

## 4. Sorting
Even though the dataframe is sorted, the original index will be kept. If you want to change or reset the index, you can use 'DataFrame.reset_index()' command.
    
    syntax
    YOUR_DATAFRAME.sort_values(by='COLUMN_NAME', ascending='False')   ascending='True' is default

```python
data_example.sort_values(by='Age') # Try ascending='False'
```

## 6. Numeric Basic functions

```python
print (data_example['Age'].min()) # Minimum value
print (data_example['Age'].max()) # Maximum value
print (data_example['Age'].mean()) # Mean value
print (data_example['Age'].std()) # Standard deviation
print (data_example['Age'].median()) # Median value
```

## 5. Merging Dataframe

```python
df_obj=pd.read_csv ('/content/drive/MyDrive/Teaching/Major Studio 1/Fall 2023/Major Stuido 1 PGDV 5200/Slides/Python_Pandas_Tutorial/data/objects.csv')
df_obj_con=pd.read_csv ('/content/drive/MyDrive/Teaching/Major Studio 1/Fall 2023/Major Stuido 1 PGDV 5200/Slides/Python_Pandas_Tutorial/data/objects_constituents.csv')
```

```python
df_obj_con_s=df_obj_con[['objectid','constituentid','displayorder',
'roletype','role','beginyear','endyear','country']]
df_obj_con_s.head()
```

[Link for pandas.DataFrame.merge Document](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html#pandas.DataFrame.merge)

DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=None, indicator=False, validate=None)

#### Logic of the HOW feature in merge
(/content/merge_index_chart.png)merge_index_chart.png

if your target column in each dataframe has different name, try the left on, right on feature.

```python
df_merge=df_obj_con_s.merge(df_con_sel, on='constituentid')
df_merge
```

## 6. Filter Rows by Criteria

```python
df_merge['roletype']=='donor'
```

```python
df_merge_drop=df_merge[df_merge['roletype']=='donor']
df_merge_drop.head()

# select the rows that fit the criteria
```

```python
df_merge_drop_new=df_merge_drop[df_merge_drop['role']!='source']
df_merge_drop_new.head()

# exclude the rows that fit the criteria
```

```python
df_merge_drop2=df_merge[(df_merge['roletype']=='donor')&(df_merge_drop['role']!='source')]
df_merge_drop2

# Or just combine them into one line
```

## 7. Grouping Data

[Link for pandas.DataFrame.groupby Document](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html)

DataFrame.groupby(by=None, axis=_NoDefault.no_default, level=None, as_index=True, sort=True, group_keys=True, observed=_NoDefault.no_default, dropna=True)

```python
df_group=df_merge_drop2[['constituentid','objectid','forwarddisplayname'
]].groupby(by=['constituentid','forwarddisplayname']).count()
df_group.head()

# can also do count, sum, mean and different calculations.
```

```python
df_group=df_group.sort_values(by=['objectid'],ascending=False)
df_group
```

Issues:
* the column names are on the different level
* no index (would cause some problems for table calculation)

```python
df_group=df_group.reset_index()
df_group
```

Issues:
* the objectid is no longer id but actually the counts

```python
df_group=df_group.rename(columns={'objectid':'counts'})
df_group
```

## 8. Export the dataset

```python
# save as csv file
df_group.to_csv(r'/content/drive/MyDrive/Teaching/Major Studio 1/Fall 2023/Major Stuido 1 PGDV 5200/Slides/Python_Pandas_Tutorial/data/groupdata.csv')

# save as json file
df_group.to_json(r'/content/drive/MyDrive/Teaching/Major Studio 1/Fall 2023/Major Stuido 1 PGDV 5200/Slides/Python_Pandas_Tutorial/data/groupdata.json',orient="records")
```

## 3. Line Chart

```python
df_obj_t=df_obj[['objectid','endyear']]
df_gr_obj_time=df_obj_t.groupby(by=
['endyear']).count().rename(columns=
{'objectid':'counts'}).reset_index()

df_gr_obj_time
```

Passing the entire dataset in long-form mode will aggregate over repeated values (each year) to show the mean and 95% confidence interval:

```python
sns.lineplot(data=df_gr_obj_time, x="endyear", y="counts")
```
